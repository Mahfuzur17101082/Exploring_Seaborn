## 3 Type of plattes

1. Circular Colors = When the data is not in ordered

```python
sns.palplot(sns.color_palette("Paired", 12))
```

2. Sequential Colors = When the data is  in ordered 

```python
sns.palplot(sns.color_palette("Blues", 12))
```

3. Diverging Colors= When both the low and high values are interesting

```python
sns.palplot(sns.color_palette("BrBG", 12))
```

```python
fid, ax=plt.subplots()
sns.distplot(t["tip"])
ax.set(xlabel="Tips")
```

```python
fid, ax=plt.subplots(2,2)
sns.distplot(t["tip"], ax=ax[0,0])
sns.distplot(t["total_bill"], ax=ax[0,1])
plt.show()
```

```python
fid, ax=plt.subplots()
sns.distplot(t["tip"])
ax.set(xlabel="Tips", xlim=(2,6), title="2 to 6 tips")
```

```python
fid, (ax0, ax1)=plt.subplots(nrows=1,ncols=2, sharey=True, figsize=(7,4))
sns.distplot(t["tip"], ax=ax0)
sns.distplot(t["total_bill"], ax=ax1)
ax0.axvline(x=np.mean(t["tip"]), label="avg tip", linestyle="--")
ax0.legend()
plt.show()
```
