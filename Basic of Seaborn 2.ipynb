## 3 Type of plattes

1. Circular Colors = When the data is not in ordered

```python
sns.palplot(sns.color_palette("Paired", 12))
```

2. Sequential Colors = When the data is  in ordered 

```python
sns.palplot(sns.color_palette("Blues", 12))
```

3. Diverging Colors= When both the low and high values are interesting

```python
sns.palplot(sns.color_palette("BrBG", 12))
```

```python
t=sns.load_dataset("tips")
print(t.shape)
t.head()
```

```python
g = sns.FacetGrid(t, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.PairGrid(t, vars=["total_bill", "tip"])
g = g.map(plt.scatter)
```

```python
g = sns.PairGrid(t, vars=["total_bill", "tip"])
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
```

```python
g = sns.pairplot(t, vars=["total_bill", "tip"], kind="reg", diag_kind="hist")
```

```python
sns.pairplot(t, vars=["total_bill", "tip", "size"], hue="smoker", palette="husl", plot_kws={'alpha':0.5}, diag_kind="hist")
```

```python
g= sns.JointGrid(data=t, x="total_bill", y="tip")
g.plot(sns.regplot, sns.distplot)
```

```python
g= sns.JointGrid(data=t, x="total_bill", y="tip")
g=g.plot_joint(sns.kdeplot)
g=g.plot_marginals(sns.kdeplot, shade=True)
```

```python
sns.jointplot(data=t, x="total_bill", y="tip", kind="hex")
```

```python
g = ( sns.jointplot(x="total_bill", y="tip", kind="scatter", xlim=(10, 40), marginal_kws=dict(bins=15, rug=True), data=t.query(' smoker == "No" ')).plot_joint(sns.kdeplot) )
```
