```python
g=sns.catplot(x="class", y="age", data=t, kind="box", col="sex")
g.fig.suptitle("Kar kemon TK", y=1.05)
g.set_titles("This is {col_name}")
g.set(xlabel="Passenger's Class", ylabel="Passenger's Age")

plt.show()
```

```python
g=sns.catplot(x="class", y="age", data=t, kind="box")
g.fig.suptitle("Kar kemon TK", y=1.05)
g.set(xlabel="Passenger's Class", ylabel="Passenger's Age")
plt.xticks(rotation=90)
plt.show()
```

Object Type: FaceGrid (ex: relpolt(), catplot()). How to Title: g.fig.subtitle()

Object Type: AxesSubplot (ex: scatterppolt(), countplot(), etc).  How to Title: g.set_title()
